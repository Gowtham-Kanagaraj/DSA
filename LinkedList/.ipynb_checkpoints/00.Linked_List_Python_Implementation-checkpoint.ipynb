```python
# Its Linear Data structure
# linked list is non - continuous memory allocation data structure
# Linked list Dynamic in nature
# Insertion / Deletion become O(n) if you know the address of the node
# Linkedlist mode up of series of connected nodes.
# where nodes contain data and the address of next node.
```

```python
# Pointers
# Pointers can store the memory address of the data
# time and memory can be less expensive when we send pointers to a function instead of variable
```

```python
# Simple Node
class Node:
    def __init__(self, data):
        self.data = data
        self.next  = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def print_linked_list(self):
        temp = self.head

        while (temp):
            print(temp.data)
            temp = temp.next
```

```python
if __name__ == '__main__':
    llist = LinkedList()
    llist.head =  Node(1)
    second = Node(2)
    third = Node(3)

    llist.head.next = second
    second.next = third

    llist.print_linked_list()
```
